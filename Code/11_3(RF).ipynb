```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import random
from torch.utils.data import TensorDataset
import matplotlib.pyplot as plt

data = pd.read_csv('/content/dataset.csv',encoding='utf-8')
data_cols=['전용면적','층','거래일자','거래금액']
data = pd.DataFrame(data[data_cols])
data.head()

```

# **RF model**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

from pylab import rcParams
import statsmodels.api as sm
import warnings
import itertools
from keras.models import Model
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split

train_feature = data.iloc[:,:-1]
train_label = data.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(train_feature, train_label, test_size=0.2)

x_train.shape, x_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor



model_rfr = RandomForestRegressor()
model_rfr.fit(x_train,y_train)
rfr_pred_y = model_rfr.predict(x_test)
```

```python
plt.figure(figsize = (21,5))
plt.plot(y_test.array, label = 'actual')
plt.plot(rfr_pred_y, label = 'prediction')
plt.title('Trade_price_Prediction',fontsize= 25)
plt.grid()
plt.legend()
plt.show()
```
