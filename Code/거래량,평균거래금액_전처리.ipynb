```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import random
from torch.utils.data import TensorDataset
import matplotlib.pyplot as plt
from copy import deepcopy

data = pd.read_csv('/content/final_apt.csv',encoding='utf-8')
data_cols=['거래일자','구','거래금액']
data = pd.DataFrame(data[data_cols])

data.head()
```

```python
data['거래일자'] = data['거래일자'].astype(int)
data.sort_values(by=['거래일자'])
start_idx = 0
last_idx = len(data)
for i in range(len(data)) :
  if data['거래일자'][i] >=20220101:
    start_idx = i
    break
print(start_idx)
```

# **2022 리스트생성**

```python
new_data = deepcopy(data[start_idx:])
new_data.head()
```

# **2022 구별 평균거래가격**

```python
mean_month = new_data.groupby('구',as_index = False)['거래금액'].mean()
mean_month
```

# **2022 구별 거래량**

```python
gu_list = [0]*16 
gu_list_labels = ['강서구','금정구','기장군','남구','동구','동래구','부산진구','북구',
                  '사상구','사하구','사하구','서구','수영구','연제구','영도구','중구','해운대구']

for i in range(start_idx,last_idx):
  for j in range(16):
    if  new_data['구'][i] == gu_list_labels[j]:
      gu_list[j]+=1

for i in range(len(gu_list)):
  print('%s의 거래량은 : %d' %(gu_list_labels[i],gu_list[i]))                 
```
